# Pandas Essentials:  Plotting

This Pandas notebook illustrates the essentials of plotting your data.  Concepts are illustrated with the [New York City pizza restaurant inspection data](https://github.com/ecerami/pydata-essentials/blob/master/pandas/data/NYC_Pizza_2017.csv), the [Iris flower data set](https://en.wikipedia.org/wiki/Iris_flower_data_set), and the [Anscombe's quartet data set](https://en.wikipedia.org/wiki/Anscombe%27s_quartet) .

Topics include:

* Histograms
* Box Plots
* Bar Plots
* Scatter Plots
* Basic Styling Options

We focus on generating "quick and dirty" plots to aid in data exploration, not beautiful, publication ready plots.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Make plots inline within Jupyter notebook
%matplotlib inline
```

```python
# To get started, we load the NYC Pizza Restaurant Inspection Data Set
pizza_df = pd.read_csv("data/NYC_Pizza_2017.csv")

# Change data type of CAMIS and ZIPCODE to String
pizza_df.CAMIS = pizza_df.CAMIS.astype(str)
pizza_df.ZIPCODE = pizza_df.ZIPCODE.astype(int).astype(str)
```

# Histograms

```python
pizza_df.hist();
```

```python
pizza_df.hist(column="SCORE");
```

```python
pizza_df.hist(column="SCORE", bins=20);
```

```python
# Kernel Density Estimate
pizza_df.plot.kde();
```

# Box Plots

```python
pizza_df.boxplot();
```

```python
pizza_df.boxplot(column="SCORE", by="BORO");
```

# Bar Plots

```python
# Bar Plots:  Pizza Restaurant Counts by Borough
counts = pizza_df.groupby("BORO").size()
counts = counts.sort_values()
counts.plot.bar();
```

```python
counts.plot.barh();
```

```python
counts.plot.pie();
```

```python
# Pizza Restaurant Counts by Business Name
counts = pizza_df.groupby("DBA").size()
counts = counts.sort_values(ascending=False)
counts = counts[0:20].sort_values()
counts.plot.barh();
```

# Scatter Plots

To illustrate scatter plots, we use the Anscombe Data Set.  You can find more details on [Wikipedia](https://en.wikipedia.org/wiki/Anscombe%27s_quartet).

```python
anscombe_df = pd.read_csv("data/anscombe.csv")
anscombe_df.head()
dataset_1 = anscombe_df[anscombe_df.dataset=="I"]
dataset_1.plot.scatter(x="x", y="y", s=100);
```

```python
dataset_2 = anscombe_df[anscombe_df.dataset=="II"]
dataset_2.plot.scatter(x="x", y="y", s=100, c="red");
```

```python
dataset_3 = anscombe_df[anscombe_df.dataset=="III"]
dataset_3.plot.scatter(x="x", y="y", s=100);
```

```python
dataset_4 = anscombe_df[anscombe_df.dataset=="IV"]
dataset_4.plot.scatter(x="x", y="y", s=100);
```
