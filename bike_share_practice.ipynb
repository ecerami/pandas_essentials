```python
two_stations = rides_df[(rides_df["Start Terminal"] == 50) & (rides_df["End Terminal"] == 56)]
```

```python
rides_df["Start Date"] = pd.to_datetime(rides_df["Start Date"])
```

```python
rides_df["DAY"] = rides_df['Start Date'].dt.day
rides_df["DAY OF WEEK"] = rides_df['Start Date'].dt.dayofweek
rides_df["MONTH"] = rides_df['Start Date'].dt.month
rides_df["DATE"] = rides_df['Start Date'].dt.date
```

```python
calendar = rides_df.groupby(by="DATE").size()
calendar_df = pd.DataFrame({"NUM_RIDES":calendar})
calendar_df.head()
```

```python
weather_df = pd.read_csv("data/babs_weather_june_july_2016.csv")
```

```python
weather_df = weather_df[weather_df.ZIP == 94107]
```

```python
weather_df["PDT"] = pd.to_datetime(weather_df["PDT"])
weather_df.index = weather_df["PDT"]
```

```python
weather_df.head()
```

```python
merged_df = calendar_df.join(weather_df)
```
